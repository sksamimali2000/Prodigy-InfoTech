```python
import numpy as np
import pandas as pd
boston_data=pd.read_csv("/content/GRadient descent Combine Cycle Power plant train Data.csv")
boston_data
```

```python
X_train=boston_data.iloc[:,0:4]
X_train
```

```python
Y_train=boston_data.iloc[:,4:5]
Y_train
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor()
model.fit(X_train,Y_train)
boston_data1=np.loadtxt("/content/G. D. Combine Cycle power plant test Data.csv",delimiter=",")
Y_predict=model.predict(boston_data1)
np.savetxt("Y_predict for boston2",Y_predict)
print("Y predict value:",Y_predict)
```
