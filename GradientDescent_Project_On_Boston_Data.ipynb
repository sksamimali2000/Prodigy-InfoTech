```python
import numpy as np
import pandas as pd
bdata=pd.read_csv("/content/Boston train Data.csv")
bdata
```

```python
X_train_boston_data=bdata.iloc[:,0:13]
X_train_boston_data
```

```python
Y_train_boston_data=bdata.iloc[:,-1]
Y_train_boston_data
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor()
model.fit(X_train_boston_data,Y_train_boston_data)
boston_data1=np.loadtxt("/content/Boston  test Data.csv",delimiter=",")
Y_predict=model.predict(boston_data1)
np.savetxt("boston_predict.csv",Y_predict)
```
